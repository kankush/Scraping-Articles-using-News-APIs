```python
# To flatten json for dict in dict
def flattenjson( json):
    val = {}
    for i in json.keys():
        if isinstance( json[i], dict ):
            get = flattenjson( json[i] )
            for j in get.keys():
                val[ i +  j ] = get[j]
        else:
            val[i] = json[i]

    return val
```
